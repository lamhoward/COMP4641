```python
# Generate a random graph
#Measure the node degree and output the matrix in adjacency matrix
```

```python
import networkx as nx
```

```python
def cal_rank(sorted_dict):
    #print (sorted_dict)
    r=0
    temp = -0.0001;
    for i,j in sorted_dict.items():

        if (sorted_dict[i] != temp): 
            r=r+1
            temp = sorted_dict[i]

        sorted_dict[i]= r
```

```python
def Req_check(dict1, dict2):
    sorted_dict1 = sorted(dict1.items(), key=lambda kv:(kv[1], kv[0]))
    sorted_dict2 = sorted(dict2.items(), key=lambda kv:(kv[1], kv[0]))
    new_dict1 = dict(sorted_dict1)
    new_dict2 = dict(sorted_dict2)
#    print(new_dict1)
#    print(new_dict2)
    cal_rank(new_dict1)
    cal_rank(new_dict2)
    for i,j in new_dict1.items():
        c1 = j;
        c2 = new_dict2[i];
        if c1 > c2: 
            return i,c1,c2
```

```python
#Q1.1
G = nx.random_partition_graph([2,3,4,5,6],0.5,0.5, seed=440)
DEG = nx.degree_centrality(G)
CLO = nx.closeness_centrality(G)
BET = nx.betweenness_centrality(G)
EIG = nx.eigenvector_centrality(G)
F = nx.to_numpy_matrix(G)
print ("Q1.1)")
print ("adajcency matrix")
print (F)
a = Req_check(DEG,CLO)
print("node_id: ", a[0])
print("closeness centrality: ", CLO[a[0]],"rank: ",a[2] ,"/20")
print("degree centrality: ", DEG[a[0]],"rank: ",a[1] ,"/20")
```

```python
#Q1.2
print ("Q1.2)")
print ("adajcency matrix")
print (F)
a = Req_check(DEG,BET)
print("node_id: ", a[0])
print("betweenness centrality: ", BET[a[0]],"rank: ",a[2] ,"/10")
print("degree centrality: ", DEG[a[0]],"rank: ",a[1] ,"/10")
```
